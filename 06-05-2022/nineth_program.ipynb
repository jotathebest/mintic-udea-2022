```python
n = int(input("Insert a number: "))

# for <variable_name> in <iterator>:
#     do something

if n > 0:
    for number in range(1, n + 1):
        print(number ** 2)
else:
    print("ERROR, number not valid")


if n <= 0:
    print("ERROR, number not valid")
else:
    for number in range(1, n + 1):
        print(number ** 2)
```
